# For initial conditions, sample the $[-1,1]^2$ rectangle uniformly

```python
t0, t1, dt = 0, 5, 0.05
t = np.arange(t0, t1, dt)
n = 100

X0_range = [geometry.sample_2d(n, [[-1,-1],[1,1]], 'random') for i in range(4)]
```

# Display state spaces under four different parameter conditions

```python
fig, ax = plt.subplots(1,4, figsize=(20,5))

def simulate_system(beta1):
    pos, vel = simulate_phase_portrait('bogdanov_takens', t, X0_range[0], par = {'beta1': -0.15, 'beta2': -0.6})
    return pos, vel

def plot_phase_portrait(pos, vel, ax):

    for p in pos:
        ax = plotting.trajectories(pos, node_feature=vel, ax=ax, style='->', lw=1, arrowhead=.03, arrow_spacing=7, axis=False, alpha=None)
    ax.axis('square')
    ax.set_xlim([-1,1])
    ax.set_ylim([-1,1])
    
pos, vel = simulate_system(-0.15)
plot_phase_portrait(ax[0])
pos, vel = simulate_system(-0.05)
plot_phase_portrait(ax[1])
pos, vel = simulate_system(0.05)
plot_phase_portrait(ax[2])
pos, vel = simulate_system(0.15)
plot_phase_portrait(ax[3])
```

# Geneate phase portraits from random initial conditions while varying $\beta_1$

```python
beta1 = np.linspace(-.15,.15,20)

pos, vel = [], []
for b1 in beta1:
    X0_range = geometry.sample_2d(n, [[-1,-1],[1,1]], 'random')
    p, v = simulate_phase_portrait('bogdanov_takens', t, X0_range, par = {'beta1': b1, 'beta2': -0.6})
    pos.append(np.vstack(p))
    vel.append(np.vstack(v))
```

# Subsample phase portrait

```python
pos_subsample, vel_subsample = [], []
for i in range(len(beta1)):
    ind, _ = geometry.furthest_point_sampling(pos[i], stop_crit=0.001)
    pos_subsample.append(pos[i][ind])
    vel_subsample.append(vel[i][ind])
    
k=5
data = utils.construct_dataset(pos_subsample, vel_subsample, graph_type='cknn', k=k)
```
